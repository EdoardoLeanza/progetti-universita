```python
import numpy as np
import math
import matplotlib.pyplot as plt

def heron(n, S, start=1):
    if n == 0:
        return start
    return 0.5 * (heron(n - 1, S, start) + (S / heron(n - 1, S, start)))
print(heron(2, 50))
print(heron(4, 50))
print(heron(8, 50))
```

```python
def heron_iteration(S, start, end):
    res = []
    for n in range(start, end):
        res.append(heron(n, S))
    return res
heron_iteration(50, 2, 9)
```

```python
def heron_plot(S, start, end, color=(0.1, 0.2, 0.5)):
    a = plt.plot(range(start, end), (heron_iteration(50, 2, 8)))
    plt.show()
    return a
heron_plot(50, 2, 8)
```

```python
def chain_heron_plot(t_start, t_end, r_start, r_end, file_name=None, to_exclude=None):
    for t in range(t_start, t_end + 1):
        if to_exclude is not None and t in to_exclude:
          continue
        heron_values = []
        for r in range(r_start, r_end + 1):
            heron_values.append(heron(r, t))

        if t - t_start < 5:
            plt.plot(range(r_start, r_end + 1), heron_values, color='#069AF3')
        else:
            plt.plot(range(r_start, r_end + 1), heron_values, color='#FFA500')

    plt.grid(True)
    plt.show()

chain_heron_plot(50, 70, 2, 8, to_exclude=range(55, 60))
```
