```python
gatto_10 = []
gatto = [1, 5, 10, 11]
for item in gatto:
  gatto_10.append(item + 10)
print(gatto_10)
```

```python
gatto_10 = [item + 10 for item in gatto]
print(gatto_10)
#codici uguali#
```

```python
gatto_11 = [item + 10 for item in gatto if item >= 10]
print(gatto_11)
```

```python
gatto_11 = []

for item in gatto:
  if item >= 10:
    gatto_11.append(item + 10)
print(gatto_11)

#codici uguali#
```

```python
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)
#numeri pari#
```

```python
even_squares = [x**2 for x in range(10) if x % 2 != 0]
print(even_squares)
#numeri dispari#
```
