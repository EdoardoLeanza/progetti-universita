```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('ChickWeight.csv')
df.head()
```

```python
df = df.iloc[:, 1:] #rimuove la prima colonna
df.describe()
#devo capire che tipo di dieta fa tipi di uova diversi --> capire la pi performante#
#chick = numero di uova#
```

```python
#average target for group#
df.groupby(['Diet']).median()
```

```python
#non possiamo dire quale sia il migliore ma ci da qualche idea sulla differenza delle 4 diete, SEMBRA che la 4 sia la migliore#
```

```python
#box plot for weight and group#
plt.boxplot(df['Diet'], df['weight'])
plt.show()
```

```python
plt.boxplot(df['Diet'])
plt.show()
#box plot che per non ci serve, ma  un inizio#
```

```python
plt.boxplot(df.groupby(['Diet']).reset_index()['weight'])
plt.show()
```

```python
w1 = df[df['Diet'] == 1]['weight']
w2 = df[df['Diet'] == 2]['weight']
w3 = df[df['Diet'] == 3]['weight']
w4 = df[df['Diet'] == 4]['weight']
plt.boxplot(w1)
plt.boxplot(w2)
plt.boxplot(w3)
plt.boxplot(w4)
plt.show()
#non va bene perch crea un unico grafico con i 4 boxplot sovrapposti#
```

```python
val = [w1, w2, w3, w4]

plt.boxplot(val)
plt.show()
#fa quello che volevamo ma nel caso in cui avessimo avuto 200 diete? direi di no#
```

```python
plt.boxplot(val)
plt.xlabel('Diet')
plt.ylabel('Weight')
plt.title('Weight Dispersion for Diets')
plt.show()
```
